```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Advertising.csv')
```

```python
df.head()
```

```python
X=df.drop('sales',axis=1)
y=df['sales']

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=101)

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(X_train)
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import ElasticNet
```

```python
base_elastic_model=ElasticNet()
```

```python
parameter_grid={'alpha':[0.1,1,5,10,50,100],'l1_ratio':[0.1,0.5,0.7,0.95,0.99,1]}
```

# GridSearchCV :-
* allows us to try multiple hyperparameters and it is not only for linear_models like ridgeCV,lassoCV it can also be used for complex models like SVM etc

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_model=GridSearchCV(base_elastic_model,param_grid=parameter_grid,scoring='neg_mean_squared_error',cv=5,verbose=2)
```

```python
grid_model.fit(X_train,y_train) # as verbose=2 give more messages
```

```python
grid_model.best_estimator_ # gives the best combination of l1_ratio and alpha
```

```python
grid_model.cv_results_ # gives bunch of information for all possible combination
```

```python
pd.DataFrame(grid_model.cv_results_)
```

```python
y_pred=grid_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test,y_pred)
```

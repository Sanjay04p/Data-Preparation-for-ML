```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('C:/Users/sanjay/Downloads/UNZIP_FOR_NOTEBOOKS_FINAL/DATA/Advertising.csv')
```

```python
df.head()
```

```python
X=df.drop('sales',axis=1)
y=df['sales']
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import Ridge
model=Ridge(alpha=100)
```

# cross_val_score: used to train and evaluate the model on different combination of datasets

```python
from sklearn.model_selection import cross_val_score
```

```python
scores=cross_val_score(model,X_train,y_train,scoring='neg_mean_squared_error',cv=5) # cv is no. of folds
```

* cv:here training data is divided into 5 folds and on each folds dataset combination changes and evaluates the model correctly 

```python
scores # these are neg_mse of all 5 folds
```

```python
abs(scores.mean()) # average of neg_mse
```

```python
model1=Ridge(alpha=1)
scores1=cross_val_score(model1,X_train,y_train,scoring='neg_mean_squared_error',cv=5)
print(scores1)
print(abs(scores1.mean()))
```

```python
model1.fit(X_train,y_train)
```

```python
y_pred=model1.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test,y_pred)
```

# cross_validate : allows us to use multiple performance metrics and also gives fit time and score time

```python
from sklearn.model_selection import cross_validate
```

```python
model=Ridge(alpha=100)
```

```python
scores=cross_validate(model,X_train,y_train,scoring=['neg_mean_squared_error','neg_mean_absolute_error'],cv=10)
```

```python
scores
```

```python
scores=pd.DataFrame(scores)
```

```python
scores
```

```python
scores.mean()
```

```python
model1=Ridge(alpha=1)
scores1=cross_validate(model1,X_train,y_train,scoring=['neg_mean_squared_error','neg_mean_absolute_error'],cv=10)
scores1=pd.DataFrame(scores1)
print(scores1.mean())
```

```python
abs(scores1['test_neg_mean_squared_error'].mean())
```

```python
model1.fit(X_train,y_train)
```

```python
y_pred=model1.predict(X_test)
```

```python
mean_squared_error(y_test,y_pred)
```

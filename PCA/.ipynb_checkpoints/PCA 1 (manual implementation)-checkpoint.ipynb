# steps to implement PCA

```python
#step1 standardize data
from sklearn.preprocessing import StandardScaler
```

```python
scalar=StandardScaler()
```

```python
std_x=scalar.fit_transform(df)
```

```python
std_x.mean()
```

```python
np.mean(std_x)
```

```python
#step2 covariance matrix
covar_matrix=np.cov(std_x,rowvar=False)
```

```python
#step3 eigen vector
eigen_values,eigen_vectors=np.linalg.eig(covar_matrix)
```

```python
eigen_vectors.shape
```

* now we have transformed 30-features of dimensions --> 30 PC space

```python
#step4 sort eigen_vectors using eigen value 
num_comp=2
```

```python
sorted_key=m=np.argsort(eigen_values) #argsort returns index position that would sort list
```

```python
sorted_key
```

```python
sorted_key[::-1][:num_comp] #grabs only largest 2PC 
```

```python
sorted_key=sorted_key[::-1][:num_comp]
```

```python
eigen_values,eigen_vectors=eigen_values[sorted_key],eigen_vectors[:,sorted_key]
```

```python
#step4 project
principle_comp=np.dot(std_x,eigen_vectors)
```

```python
principle_comp
```

```python
plt.scatter(principle_comp[:,0],principle_comp[:,1])
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('C:/Users/sanjay/Downloads/UNZIP_FOR_NOTEBOOKS_FINAL/DATA/cancer_tumor_data_features.csv')
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar=StandardScaler()
```

```python
std_x=scalar.fit_transform(df)
```

```python
from sklearn.decomposition import PCA
```

```python
model=PCA(n_components=2)
```

```python
model.fit(std_x) #calculate eigen vectors and eigen values
```

```python
principle_comp=model.transform(std_x) #projecting data
```

* fit_transform can be used to get results in one step

```python
plt.scatter(principle_comp[:,0],principle_comp[:,1])
```

```python
model.components_
```

```python
df_comp=pd.DataFrame(model.components_,index=['PCA1','PCA2'],columns=df.columns)
```

```python
df_comp
```

```python
plt.figure(figsize=(20,3))
sns.heatmap(df_comp,annot=True)
```

```python
model.explained_variance_ratio_
```

```python
np.sum(model.explained_variance_ratio_)
```

* here 44+18=63% of variance is explained by these above two components

# choosing a best n_components for PCA

```python
explained_variance=[]

for i in range(1,30):
    pca=PCA(n_components=i)
    pca.fit(std_x)
    
    explained_variance.append(np.sum(pca.explained_variance_ratio_))
```

```python
plt.plot(range(1,30),explained_variance)
plt.xlabel("Number of components")
plt.ylabel("Variance Explained")
```
